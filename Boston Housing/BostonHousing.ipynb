<b><b>      Robust Regression using Random Sample Consensus</b></b>

```python
X = df['RM'].values.reshape(-1,1)
y = df['MEDV'].values
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X, y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
```

```python
np.arange(3,10,1)
```

```python
line_X = np.arange(3,10,1)
line_y_ransac = ransac.predict(line_X.reshape(-1,1))
```

```python
sns.set(style='darkgrid', context='notebook')
plt.figure(figsize=[12,8])
plt.scatter(X[inlier_mask], y[inlier_mask],
           c='blue', marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
           c='brown', marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='red')
plt.xlabel('average number of rooms per dwelling')
plt.ylabel("Median Value of owner-occupied homes in $1000's")
plt.legend(loc='upper left')
plt.show()
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

<b>What a perfect model should look like !!<b>

```python
generate_random = np.random.RandomState(0)
```

```python
x = 10*generate_random.rand(1000)
y = 3*x + np.random.randn(1000)
```

```python
plt.figure(figsize = (10,0))
```

```python
plt.scatter(x, y)
```

```python
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)
```

```python
model.fit(X_train.reshape(-1, 1), y_train)
```

```python
y_train_pred = model.predict(X_train.reshape(-1, 1))
y_test_pred = model.predict(X_test.reshape(-1, 1))
```

<b>Method 1 : Residual Analysis</b>

The residual plot shows a fairly random pattern - the first residual is positive, the next two are negative, the fourth is positive, and the last residual is negative. This random pattern indicates that a linear model provides a decent fit to the data.



```python
plt.figure(figsize=(12,10))
plt.scatter(y_train_pred, y_train_pred - y_train, c='blue', marker='o', label='Training data')
plt.scatter(y_test_pred, y_test_pred - y_test, c='orange', marker='*', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-3, xmax=33, lw=2, color='k')
plt.xlim([-5, 35])
plt.ylim([-25, 15])
plt.show()
```

<b>Method 2: Mean Squared Error (MSE)</b>

The MSE is a measure of the quality of an estimatorit is always non-negative, and values closer to zero are better.

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_train, y_train_pred)
mean_squared_error(y_test, y_test_pred)
```

<b>Mthod 3: Coefficient of Determination, R2</b>

A value of 1.0 indicates a perfect fit, and is thus a highly reliable model for future forecasts, while a value of 0.0 would indicate that the calculation fails to accurately model the data at all. But a value of 0.20, for example, suggests that 20% of the dependent variable is predicted by the independent variable

```python
from sklearn.metrics import r2_score
r2_score(y_train, y_train_pred)
r2_score(y_test, y_test_pred)
```
