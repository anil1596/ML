# Getting Started


# Problem Statement

- Something simple to understand to rantionale of using the neural Networks.
- We have some sample data generated by the following fn: **y = 3X + 50**
- However in the real life we have no knowledge of the function. We only have the data and have to estimate the function with the data given. that is the challenge here.

## Import Libraries

```
import numpy as np

# install Tenserflow
try: 
  %tensorflow_version 2.x
except Exception:
  pass

import tensorflow as tf
```

## Preparing the Training Data

```
X = np.array([-20, -5, 1, 5, 20, 55, 76], dtype=float)
y = np.array([-10, 35, 53, 65, 110, 215, 278], dtype=float)
```

### Side Note
- Feature - Independent variable(s)/input(s) to our model.
- Label/Target(s) - The dependent variable(s)/output of our model.
- Example - A pair of feature/label provided to the machine learning algorithm during training.

### Defining Our model
- We'll start with only one **neuron** and one dense layer

#### Model and Layer Specifications
 We need to first instantiate (create an instance) layer by calling *tf.keras.layers.Dense*

 - input_shape=[1] - Value that we are only taking a single value to this layer. In our case, X.

 - units=1 - Value of 1 specify that we are informing keras we are creating a neural network with only one *neuron* (some time called internal variable). The more complex the problem is the more neurons you will need to learn the pattern. May need more than a layer as well.

 - For our problem we are taking only a single layer. (same as final layer = 1). The units specify the number of output from our neural network- a single value (single floating point number)

- For multilayered network, the size and shape of the ouput layer as defined by units would need to match the *input_shape* of next layer.

```
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

# Compile Model

 - **Loss function** - the loss function is a performance metric. it is used to measure the errors of our model against the actual outcome. It is called loss fn. because the difference between our predicion and actual is called "loss"
 - **optimizer function** - mathematical fn. that is used to adjust model parameters to improve the prediction accuracy.

```
learning_rate = 0.1 # 0.1 is faster than 0.01 => specifies the step size taken by optimizer when adjusting values in the model
```

```
model.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.Adam(learning_rate))
```

# Training Model

- Neural Network starts off with the random value (also called the weights) and use it adn the input X to compute y. As weights started off with a random value, the predictions will start off with large errors.
- loss calue is calculated using the lasses fn. taking the prediction and acutal value into account.
- optimize then make slight (acc. to learning rate) adjustments to the weights with the goal of reducing the loss.
- the **epoch** is the full iteration of the examples we have seen i.e. all our data.

## Parameters
- input X 
- output y
- epoch specifies how many cycles should be run
- verbose argument controls how much output the fit method is. displayed on the console.


```
history = model.fit(X, y, epochs=1000, verbose=False)
```

# Visualising the Model Training

```
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel('Loss Magnitude')
plt.plot(history.history['loss'])
```
